## import libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## creating training set

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
training_set = train_datagen.flow_from_directory(
    'dataset/training_set',
    target_size=(64,64),
    batch_size = 32,
    class_mode = 'binary'
)
```

## create test set

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
    'dataset/test_set',
    target_size = (64,64),
    batch_size = 32,
    class_mode = 'binary'
)
```

## make da cnn

```python
cnn = tf.keras.models.Sequential()
```

## input shape and convolutional layer + pooling

```python
cnn.add(tf.keras.layers.Input(shape = (64,64,3)))
cnn.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides= 2))
```

## next layer

```python
cnn.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides= 2))
```

## flatten

```python
cnn.add(tf.keras.layers.Flatten())
```

## connected layer

```python
cnn.add(tf.keras.layers.Dense(units = 120, activation = 'relu'))
```

## output layer

```python
cnn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

## compile

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## train the cnn

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

## preprocess the image to predict & predict

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/dog.png', target_size = (64,64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)

result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
    prediction = 'dog'
else:
    prediction = 'cat'
```
