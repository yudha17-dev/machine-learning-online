# Thompson Samping

## Importing libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

## Implementing Thompson Sampling

```python
import random
N = 500
d = 10
ads_selected= []
number_of_reward = [0] * d
number_of_fail = [0] * d
total_reward = 0
for n in range(0,N):
    max_theta = 0
    ad = 0
    for i in range(0,d):
        theta = random.betavariate(number_of_reward[i]+1,number_of_fail[i]+1)
        if theta > max_theta:
            ad = i
            max_theta = theta
    ads_selected.append(ad)
    reward = dataset.values[n,ad]
    total_reward += reward
    if reward == 1:
        number_of_reward[ad] += 1
    else:
        number_of_fail[ad] += 1
```

## Plotting the histogram

```python
plt.hist(ads_selected)
plt.title('Histogram of Ads')
plt.xlabel('this')
plt.ylabel('that')
plt.show()
```
